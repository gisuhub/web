```python
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"],},
                    index=[0,1,2,3])
```

```python
df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                       "B":["B4","B5","B6","B7"],
                       "C":["C4","C5","C6","C7"],
                       "D":["D4","D5","D6","D7"],},
                        index=[4,5,6,7])
```

```python
left = pd.DataFrame({"key":["K0","K1","K2","K3"],
                    "A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"]})
right = pd.DataFrame({"key":["K0","K1","K2","K4"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]})
```

```python
left
```

```python
right
```

```python
# 데이터 Merging: 동일한 행이나 형을 기준으로 내용의 데이터셋을 합할 수 있는 메소드
# Left와 right 두가지 데이터 중 내용이 같은 key를 기준으로 두 데이터를 합쳐서 보여줌. 
# Inner라는 조건은, 두개의 데이터 셋에 공통된 부분만 합하여 준다는 뜻.
pd.merge(left,right,how='inner',on='key')
```

```python
right
```

```python
pd.merge(left,right,how='outer',on='key')
# 합치려는(Merge 하려는) 데이터셋의 key를 모두 가져와서, 
# 기준 컬럼으로 사용하고, 해당 key에 데이터가 없는 부분은 NaN으로 출력되도록 한다. 
```

```python
left1 = pd.DataFrame({"key1":["K0","K0","K1","K2"],
                      "key2":["K0","K1","K0","K1"],
                    "A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"]})
right1 = pd.DataFrame({"key1":["K0","K1","K1","K2"],
                       "key2":["K0","K0","K0","K0"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]})
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1,right1, on=['key1','key2'])
```

```python
pd.merge(left1,right1,how="outer", on=['key1','key2'])
```

```python
pd.merge(left1,right1,how="left", on=['key1','key2'])
```
