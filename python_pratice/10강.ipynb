```python
#Plt에 matplotlib 을 저장했고, matplotlib에 있는 plot이라는 메소드를 통해 어떤 상관관계가 있는 
#그래프를 그릴 것인지를 지정하고, 해당 그래프의 색깔을 지정한다. 
#plot(x축에 대입할 값, y축에 대입할 값, ‘라인 컬러’)
#Plot : 이야기 
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.3,0.9])

axes.plot(x,y,'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

# Figure라는 메소드는 그래프의 크기와 가로 세로 비율을 정할 수 있도록 해준다. 
# fig.add_axes([왼쪽축에서의 거리, 바닥에서의 거리, 그래프의 폭, 그래프의 높이]) 
# 위 add_axes 안에 들어갈 수 있는 숫자의 범위는 0부터 1까지 가능 


```python
fig = plt.figure()
axes2 = fig.add_axes([0.1,0.1,0.9,0.9])
axes3 = fig.add_axes([0.2,0.6,0.4,0.3])

axes2.plot(x,y,'b')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')

axes3.plot(y,x,'r')
axes3.set_xlabel('X_label_axes2')
axes3.set_ylabel('Y_label_axes2')
axes3.set_title('Axes 2 Title')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.9,0.8])

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.tick_params(direction = 'in')
```
