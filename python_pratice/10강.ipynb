```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
#X 를 x축에, y를 y축에 대입하고, ‘r’ 은 그래프의 색깔을 red로 표현해라.
```

```python
plt.plot(y,x,'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String Title Here')
plt.show()
```

```python
plt.plot(x,y,'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String Title Here')
plt.show()
```

```python
plt.plot(x,y,'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String Title Here')
plt.show()
```

```python
plt.plot(x,y,'y')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String Title Here')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.subplot(1,3,3)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(2,3,2)
plt.plot(x,y,'r--')
plt.subplot(2,3,4)
plt.plot(y,x,'g*-')
```

```python
#plt.subplot(1,2,1)의 의미는, 기존 plot을 1을 기준으로 
# 가로에 두개의 그래프를 넣는다는 의미이고, 그 다음 1은 두개의 
#그래프 위치 중 첫번째에 plt.plot(x, y, 'r--')  라는 그래프를 넣는다는 의미이다.
#plt.subplot(1,3,1) 은 기존 plot의  범위에 3개의 그래프를 넣는다는 의미이고, 
#그 그래프 위치 3가지 중 첫번째에 plt.plot(x, y, 'r--') 그래프를 넣는다는 뜻이다. 
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,'y--')
plt.subplot(1,4,4)
plt.plot(y,x,'b*-')
```
