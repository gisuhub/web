# Bar Plots

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]

plt.figure(figsize=(6,6))
plt.bar(plants,germinated, color='orchid')
plt.ylabel('Number Of Seeldings')
plt.title('Germination Rates')
```

```python
plt.figure(figsize=(6,6))

plt.bar(plants,died , color='red')
plt.ylabel('Number Of Seeldings')
plt.title('Failure Rates')
```

```python
plt.figure(figsize=(6,6))

plt.bar(plants,died, color='#3498db')
plt.bar(plants,germinated , color='#2ecc71',width=0.4, align='edge')

plt.ylabel('Number Of Seeldings')
plt.title('Failure Rates')
```

```python
plt.figure(figsize=(6,6))

plt.bar(plants,died, color='#3498db',bottom=germinated, label ='Failed Seedlings')
plt.bar(plants,germinated , color='#2ecc71',label = 'Germinated Seedlings')
plt.legend()
plt.title('Greenhouse Seedlings')
```

```python
plt.figure(figsize=(6,6))

plt.bar(plants,died, color='#3498db',bottom=germinated, label ='Failed Seedlings')
plt.bar(plants,germinated , color='#2ecc71',label = 'Germinated Seedlings')
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title('Greenhouse Seedlings')
```

```python
import pandas as pd
titanic = pd.read_csv('data/titanic.csv')

titanic['age'] = pd.to_numeric(titanic['age'], errors='coerce')
titanic['pclass'] = pd.to_numeric(titanic['pclass'], errors='coerce')
```


Errors = “coerce” 라고 표기하면 숫자가 아닌 데이터를 NaN으로 설정하여 반환한다. 
Pd.to_numeric 은 데이터를 숫자 형식으로 표출하라는 것. 

Age를 기준으로 range에 해당하는 행의 갯수를 더한 결과가 아래 그래프로 표기 되어 있다

```python
plt.hist(titanic['age'])
```

```python
plt.hist(titanic['pclass'])
```

```python
heights = [137,140,142,145,147,150,152,155,157,160]
f_weights = [28.5,30.8,32.6,34.9,36.4,39,40,43.1,44.9,47.2]
m_weights = [34.9,38.1,33.5,35.8,46.7,42.8,43.1,45.8,50.8,58.9]
```

```python
import matplotlib.pyplot as plt


plt.plot(heights,f_weights, marker = '*', label = 'Female')
plt.plot(heights,m_weights, marker = 'd', label = 'male')

plt.ylabel('Weight (kg)')
plt.xlabel('Height (cm)')

plt.legend()
```

남녀의 몸무게가 키에 따라 어떻게 상관관계를 갖는지 그래프로 드러내고자 함. 
여자의 경우 키에 따른 몸무게의 변화가 일정하게 늘어나지만, 
남자의 경우, 140~ 145구간과 147에서 155구간이 키에 비해 저체중 양상이 나타납니다. 




```python
plt.scatter(heights,f_weights, marker = '*', label = 'Female')
plt.scatter(heights,m_weights, marker = 'd', label = 'male')

plt.ylabel('Weight (kg)')
plt.xlabel('Height (cm)')

plt.legend()
```

```python
first_class = titanic[titanic['pclass']==1]['age']
second_class = titanic[titanic['pclass']==2]['age']
third_class = titanic[titanic['pclass']==3]['age']

plt.figure(figsize=(10,4))
ax = plt.subplot(1,3,1)
plt.hist(first_class,label = '1st Class', color='#f1c40f',bins=50)
plt.title('1st Class')
plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class,label = '2st Class', color='blue',bins=50)
plt.title('2nd Class')
plt.subplot(1,3,3, sharey=ax)
plt.hist(third_class,label = '3st Class', color='red',bins=50)
plt.title('3rd Class')
```

Bins값을 높이면 높일수록 데이터의 정확도가 높아지고 세밀해집니다. X축 간격이 좋아집니다.
